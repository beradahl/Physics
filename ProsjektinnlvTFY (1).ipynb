# Oppgave 1
Ut fra plot ser man at de numerisk beregnede energiverdiene stemmer med de som den analystiske løsningen gir oss for forskjellige k verdier, og at bølgefunksjonene har lik form.

Integraltesten viser at funksjonen er normert og sumtesten viser den er ortogonal. De danner et fullstendig sett fordi de er egenfunksjoner som er lineært uavhengige.

```python
import matplotlib.pyplot as plt
import numpy as np

dx=1e-11
hbar=1.05e-34
m=9.11e-31
N=100
V = [0 for n in range(N)]

d = [((hbar**2)/(m*dx**2)) + v for v in V]
e = -(hbar**2)/(2*m*dx**2)

H = [[0]*(N) for n in range(N)]
for i in range(N):
    for j in range(N):
        if i==j:
            H[i][j]=d[i]
        if abs(i-j)==1:
            H[i][j]=e
          
w, v = np.linalg.eigh(H)

bølg1 = v[:,0]
bølg2 = v[:,1]
bølg3 = v[:,2]
bølg4 = v[:,3]
bølg5 = v[:,4]

x = [dx*1e9*n for n in range(N)]
plt.plot(x,bølg1,x,bølg2,x,bølg3,x,bølg4,x,bølg5)
plt.title('partikkel i boks',fontsize=14)
plt.xlabel('$x$',fontsize=14)
plt.ylabel('$\psi$',fontsize=14)
plt.show()

print("Energiegenverdier:")
print('E1=', w[0])
print('E2=', w[1])
print('E3=', w[2])
print('E4=', w[3])
print('E5=', w[4])

#integraltest
y = np.linspace(0,1.0*N,N)
g1=np.trapz(np.abs(bølg1)**2, y)
print("Er normert hvis integralpsi^2=1, integral psi^2=",g1)

#ortotest
j1=np.sum(bølg1**2)
print("Er ortogonal hvis sum $\psi^2$=1, sum psi =", j1)
```

# Oppgave3
Energien til to separate atomer er 2* energien til grunntilstanden til et atom (hentet fra oppgave 2), mens energien til molekylet er 2* energien til grunntilstanden til et molekyl. Bindingsenergien blir da Eatom-Emolekyl = 3.527254026297725e-29.
Denne bindingsenergien er velidg lav i forhold til virkeligheten.

```python
import matplotlib.pyplot as plt
import numpy as np

hbar=1.05e-34
m=9.11e-31
V0=-2*1.6E-19
N=100
dx=1e-11

side1 = [0]*2*N
atom1 = [V0]*N
mellom = [0]*3*N
atom2 = [V0]*N
side2 = [0]*2*N
V = side1+atom1+mellom+atom2+side2

d = [(hbar**2/(m*dx**2))+ v for v in V]
e = -(hbar**2/(2*m*dx**2))

sumN=len(V)

H = [[0]*(sumN) for n in range(sumN)]
for i in range(sumN):
    for j in range(sumN):
        if i==j:
            H[i][j]=d[i]
        if abs(i-j)==1:
            H[i][j]=e
         
w,v = np.linalg.eigh(H)

x = [dx*n for n in range(sumN)]

bølg1 = v[:,0]
bølg2 = v[:,1]
bølg3 = v[:,2]
bølg4 = v[:,3]

plt.plot(x,bølg1,x,bølg2,x,bølg3,x,bølg4)
plt.title('dobbel potensialbrønn molekyl 1D',fontsize=14)
plt.xlabel('$x$',fontsize=14)
plt.ylabel('$\psi$',fontsize=14)
plt.show()

print('Energiverdier')
print('E1=',w[0])
print('E2=',w[1])
print('E3=',w[2])
print('E4=',w[3])

plt.plot(x,V)
l = plt.axhline(y=w[0], linewidth=1, color='b')
l = plt.axhline(y=w[1], linewidth=1, color='g')
l = plt.axhline(y=w[2], linewidth=1, color='r')
l = plt.axhline(y=w[3], linewidth=1, color='c')
plt.title('energinivå dobbel potensialbrønn molekyl',fontsize=14)
plt.xlabel('$x$',fontsize=14)
plt.ylabel('$E$ (J)',fontsize=14)
plt.show()

#bindingsenergi
Emolekyl = 2*w[0]
Eatom = 2*-2.835900344968736e-19 #fra oppgave 2

bindings = Eatom-Emolekyl
print('bindingsnergi=', bindings)
```
