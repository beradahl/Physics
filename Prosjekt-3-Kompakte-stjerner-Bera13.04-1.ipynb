### c)
Vi reduserer Lane-Emden til to første-ordens differensiallikninger ved å innføre at $\frac{d\theta}{d\xi}=\chi$. Hvis vi først skriver om til $$\frac{1}{\xi^2}(\frac{d}{d\xi}(\xi^2)\frac{d\theta}{d\xi}+(\xi^2\frac{d\theta}{d\xi}\frac{d\theta}{d\xi}))=-\theta^n$$ $$\Rightarrow \frac{1}{\xi^2}(2\xi \frac{d\theta}{d\xi}+\xi^2\frac{d^2\theta}{d\xi^2})=-\theta^n$$ $$\Rightarrow \frac{2}{\xi}\frac{d\theta}{d\xi}+\frac{d^2\theta}{d\xi^2})=-\theta^n$$ Da kan vi sette inn $\chi$: $$\frac{2}{\xi}\chi+\frac{d\chi}{d\xi}=-\theta^n$$ slik at den andre diff-likningen blir $$\frac{d\chi}{d\xi}=-\theta^n-\frac{2}{\xi}\chi$$ slik får vi $$\frac{d \overrightarrow{\omega}}{d\xi} = f\left(\overrightarrow{\omega}\right) = f\begin{bmatrix} \theta \\ \chi \end{bmatrix} = \begin{bmatrix} \chi \\ -\theta^n - \frac{2\chi}{\xi} \end{bmatrix}$$


```python
#c) Lane-Emden skrevet på vektorform 
import numpy as np
import matplotlib.pyplot as plt

def f(w, xi, n):
    """Definerer Lane-Emden som to første-ordens diff.likningen
       w er initialverdiene til theta og x på formen w=[theta,x],
       n er hvilken n vi skal løse Lane-Emden for.
       Returnerer en approksimasjon av Lane-Emden
    """
    approks_w = np.zeros(len(w))
    approks_w[0] = w[1]
    if w[0] < 0:
        approks_w[1] = - 2/xi * w[1]
    else:
        approks_w[1] = -(w[0])**n - 2/xi * w[1]

    return approks_w
```

### d)
Implementasjon i Eulers:
$$\theta_{n+1} = \theta_n + \chi\cdot\Delta\xi$$ $$$$
$$\chi_{n+1} = \chi_n - \left(\theta_{n}^n + \frac{2\chi}{\xi}\right)\cdot \Delta\xi$$

```python
#d) Eulers ett steg
def Euler_steg(w, h, f, xi, n):
    """Tar ett steg med Eulers ved xi, der w er initialverdiene, 
        h er steglengden, f er funksjonen som skal approksimerers.
        Returnerer en ny vektor w med nye verdier for f.
    """
    ny_w = w + h * f(w, xi, n)
    return ny_w
```

```python
#Euler for hele greia
def Euler(f, w, x_0, x_N , n, h):
    """Tar N steg med funksjonen f
       xi_0 er startverdien til xi, xi_N er der theta=0.
       Returnerer en matrise med theta og X, og verdiene for xi.
       """
    
    N=int((x_N-x_0)/h) #oppgitt formel for N
    
    ordn = len(w)
    
    Matrise = np.zeros((ordn, N)) #lager tom matrise
    Matrise[:,0] = w #setter initalverdiene til theta og X inn i matrisen

    x_verdier=np.linspace(x_0, x_N, N-1) #lager xi-verdiene, bruker N-1 punkter for å forsikre oss om at vi får med xi_N
    
    
    for i in range(1,N):
        Matrise[:,i] = Euler_steg(Matrise[:, i-1], h, f, x_verdier[i-1], n) #regner og legger neste steg inn i matrisen
        
    x_verdiene = np.insert(x_verdier,0,0) #legger til at den første xi=0
    return Matrise, x_verdiene
```

```python
xi_N=np.pi

thetaer1, xi1 = Euler(f,[1,0],0.01,xi_N,1,0.1) #regnet med h=0.01
thetaer3, xi3 = Euler(f,[1,0],0.01,xi_N,1,0.005) #regnet med h=0.005
thetaer4, xi4 = Euler(f,[1,0],0.01,xi_N,1,0.0001) #regnet med h=0.0001
```

```python
#Det analytiske uttrykket for theta når n=1

def analytisk_theta(xi):
    return np.sin(xi)/xi

x=np.linspace(0.01,xi_N)

lane_theta=[0]*len(x)

for i in range(0,len(x)):
    lane_theta[i]=(analytisk_theta(x[i]))
```

```python
"""Plotting av analytiske og numeriske theta-verdier ved Euler"""

plt.rcParams.update({'figure.figsize':(11.0, 6.0)})

plt.plot(x,lane_theta,label = 'analytisk')
plt.plot(xi1,thetaer1[0,:],linestyle = '-.', label = 'Euler med h=0.01')
plt.plot(xi3,thetaer3[0,:],linestyle = ':', label = 'Euler med h=0.005')
plt.plot(xi4,thetaer4[0,:],linestyle = '-.', label = 'Euler med h=0.0001')
plt.ylabel(r'$\theta(\xi)$',fontsize = '14')
plt.xlabel(r'$\xi$',fontsize = '14')
plt.title(r'$\theta(\xi)$ analytisk og numerisk ved n=1 med Euler')
plt.legend()
plt.grid(True)
plt.show()

print("Theta_N for h=0.01", thetaer1[0,:][-1])
print("Theta_N for h=0.005", thetaer3[0,:][-1])
print("Theta_N for h=0.0001", thetaer4[0,:][-1])
```

```python
# f) Rungekutta

def rungestep(w,xi,h,f,n):
    """Definerer ett steg med Runge-Kutta metoden"""
    
    k_1 = f(w, xi, n)
    k_2 = f(w + k_1 * h/2, xi + h/2, n)
    k_3 = f(w + k_2 * h/2, xi + h/2, n)
    k_4 = f(w + k_3 * h, xi + h, n)
    next_w = w + h/6* (k_1 + 2 * k_2 + 2 * k_3 + k_4)
    return next_w
    

    
def Runge(f, w, xi, xi_N , n, h): #Hele Runge-Kutta
    N = int(abs(xi_N-xi)/h) #ant_steps
    xvals = np.linspace(xi,xi_N,N-1)
    
    deg = len(w)
   
    wvals = np.zeros((deg,N))
    wvals[:,0] = w
   
    for i in range(1,N):
        wvals[:,i] = rungestep(wvals[:, i-1],xvals[i-1],h,f,n)
    xvalues = np.insert(xvals, 0, 0)
    return wvals, xvalues
```

```python
#f) utregning av numerisk ved Runge-Kutta
wvals1, xvals1 = Runge(f, w, 0.001,np.pi,1, 0.1)
wvals2, xvals2 = Runge(f, w, 0.001,np.pi,1, 0.05)
wvals3, xvals3 = Runge(f, w, 0.001,np.pi,1, 0.001)
wvals4, xvals4 = Runge(f, w, 0.001,np.pi,1, 0.0001)
```

### h)
Radien av stjerna er bestemt ved at trykket på overflaten er lik 0, noe som da betyr at trykket mot overflaten må avta. Dermed er trykket høyest i senteret til stjernen, der r = 0. Fra oppgave 2b) har vi at $$\overline{P}(x)=\frac{\sqrt{1-\alpha}-\sqrt{1-\alpha x^2}}{\sqrt{1-\alpha x^2}-3\sqrt{1-\alpha}}$$ For $x=0$ vil vi da få $$\overline{P}(0)=\frac{\sqrt{1-\alpha}-\sqrt{1-\alpha 0^2}}{\sqrt{1-\alpha 0^2}-3\sqrt{1-\alpha}}$$ $$\Rightarrow \overline{P}(0)=\frac{\sqrt{1-\alpha}-1}{1-3\sqrt{1-\alpha}}$$ Ser vi på $\frac{d\overline{P}}{dx}$ er det tydelig at trykket avtar, da  $$\frac{d \overline{P}}{dx}=-\frac{1}{2} \alpha x[1+\overline{P}][1+3\overline{P}][1-x^2\alpha]^{-1}$$ slik at funksjonens stigningstall er negativt. Ved $x=0$ får vi  $$\frac{d\overline{P}}{dx}|_{x=0} = 0$$

```python
def g(w, x, a):
    """Definerer TOV-likningen
       w er initialverdien til P,
       a er hvilken a vi skal løse TOV for.
       Returnerer en numerisk approksimasjon av TOV
    """
    return(-1/2*a*x*(1+w[0])*(1+3*w[0]))/(1-a*x**2)
```

```python
def P_bar(x, a,w):
    return (np.sqrt(1-a)-np.sqrt(1-a*x**2))/(np.sqrt(1-a*x**2)-3*np.sqrt(1-a))
```

```python
"""Regner ut analytisk og numerisk TOV for Euler og RK4 for forskjellige alpha-verdier, 
   her har vi valgt og fortsette å bruke en steglengde på h=0.0001 både for Euler og RK4"""

xx=np.linspace(0,1)

#for alpha=0.86
w1=[(np.sqrt(1-0.86)-1)/(1-3*np.sqrt(1-0.86)),0]
P1=P_bar(xx,0.86,w1)

TOV1E, x1E = Euler(g, w1, 0.001, 1 , 0.86, 0.0001)
TOV1R, x1R = Runge(g, w1, 0.001, 1 , 0.86, 0.0001)

#for alpha=0.59
w2=[(np.sqrt(1-0.59)-1)/(1-3*np.sqrt(1-0.59)),0]
P2=P_bar(xx,0.59,w2)

TOV2E, x2E = Euler(g, w2, 0.001, 1 , 0.59, 0.0001)
TOV2R, x2R = Runge(g, w2, 0.001, 1 , 0.59, 0.0001)

#for alpha=0.0167
w3=[(np.sqrt(1-0.0167)-1)/(1-3*np.sqrt(1-0.0167)),0]
P3=P_bar(xx,0.0167,w3)

TOV3E, x3E = Euler(g, w3, 0.001, 1 , 0.0167, 0.0001)
TOV3R, x3R = Runge(g, w3, 0.001, 1 , 0.0167, 0.0001)
```
